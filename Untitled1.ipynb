```python
rdf = dfk.groupby('ts')['temp'].median()
rdf.plot()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = rdf.index
y = rdf.values


a1 = np.polyfit(x, y, 1)
f1 = np.poly1d(a1)
f1(2050)
rdf.plot()
plt.plot(x,f1(x))
```
