```python
#%m-%d %H:%M:%S
#dfk = df[df['device']=='b8:27:eb:bf:9d:51'][['ts','humidity','temp']]
dfk = df[df['device']!=''][['humidity','temp','co','lpg','smoke']]
#dfk = df[df['device']!=''][['humidity','temp']]
#dfk.head()

#dfk['monf']=df['ts'].apply(lambda x: int(time.strftime('%m',time.localtime(x))))
dfk['day']=df['ts'].apply(lambda x: int(time.strftime('%d',time.localtime(x))))
#dfk['Hour']=df['ts'].apply(lambda x: int(time.strftime('%H',time.localtime(x))))
#dfk['Mins']=df['ts'].apply(lambda x: int(time.strftime('%M',time.localtime(x))))
#dfk['Sec']=df['ts'].apply(lambda x: int(time.strftime('%S',time.localtime(x))))
#dfk['dtis']=df['ts'].apply(lambda x: int(int(x)-time.mktime(time.strptime(str(time.strftime('%y-%m-%d',time.localtime(x))), '%y-%m-%d'))))
dfk['dtis']=df['ts'].apply(lambda x: int(int(x)-time.mktime(time.strptime(str(time.strftime('%y-%m-%d',time.localtime(x))), '%y-%m-%d')))/1000)
#dfk['light']=df['light'].apply(lambda x: 1 if x==True else 0)
dfk['motion']=df['motion'].apply(lambda x: 1 if x==True else 0)
dfk['co']=df['co'].apply(lambda x: x*1000)
dfk['lpg']=df['lpg'].apply(lambda x: x*1000)
dfk['smoke']=df['smoke'].apply(lambda x: x*1000)


```

```python
#Pdict = {'Urban': 0, 'Samyurban': 1, 'Nourban': 3}
#dfk['столбец']=df['столбец'].apply(lambda x: Pdict[x])
#Pdict
Pdict = {True: 'True', False: 'False'}
dfk['light']=df['light'].apply(lambda x: Pdict[x])

Pdict = {hash("True"): 1, hash("False"): 0}
dfk['light']=df['light'].apply(lambda x: hash(x))
dfk.head()
dfk['light']=dfk['light'].apply(lambda x: Pdict[x])
#Pdict
dfk.head()
```
