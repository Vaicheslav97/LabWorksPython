```python
import warnings 
warnings.filterwarnings('ignore')

#%matplotlib inline
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LogisticRegression, LogisticRegressionCV
from sklearn.model_selection import cross_val_score, StratifiedKFold
from sklearn.model_selection import GridSearchCV
```

```python
#sns.lmplot('co','temp',data =dfk, hue='light', fit_reg=False);

plt.scatter(Data_train[Marks_train== 0][['blue']],Data_train[Marks_train== 0][['red']])
plt.scatter(Data_train[Marks_train== 1][['blue']],Data_train[Marks_train== 1][['red']], c='Orange')
```

```python
def plot_boundary(clf, X, Y, grid_step=.01, poly_featurizer=None):
    x_min, x_max= X.min()-.1, X.max()+.1
    y_min, y_max= Y.min()-.1, Y.max()+.1
    xx,yy = np.meshgrid(np.arange(x_min, x_max, grid_step),
                        np.arange(y_min, y_max, grid_step))
    Z =clf.predict(poly_featurizer.transform(np.c_[xx.ravel(),yy.ravel()]))
    Z = Z.reshape(xx.shape)
   # plt.contour(xx,yy,Z,cmap=plt.cm.paired)
    plt.contour(xx,yy,Z,cmap=plt.cm.get_cmap("Spectral"))
```

```python
poly = PolynomialFeatures(degree=7)
X_poly = poly.fit_transform(Data_train[['blue','red']],y=None)
```

```python
C= 1e-2
logit = LogisticRegression(C=C, random_state=17)
logit.fit(X_poly, y =Marks_train)
#'blue','red'
#plot_boundary(logit,Data_train['blue'],Data_train['red'],grid_step=.01, poly_featurizer= poly)

#plt.scatter(Data_train[Marks_train== 0][['blue']],Data_train[Marks_train== 0][['red']])
#plt.scatter(Data_train[Marks_train== 1][['blue']],Data_train[Marks_train== 1][['red']], c='Orange')
#plt.title('Logit with C=%s' % C)

#print("Accuracy on training set:", round(logit.score(X_poly,Marks_train),3))
```

```python
plot_boundary(logit,Data_train['blue'],Data_train['red'],grid_step=.01, poly_featurizer= poly)

plt.scatter(Data_train[Marks_train== 0][['blue']],Data_train[Marks_train== 0][['red']])
plt.scatter(Data_train[Marks_train== 1][['blue']],Data_train[Marks_train== 1][['red']], c='Orange')
plt.title('Logit with C=%s' % C)

print("Accuracy on training set:", round(logit.score(X_poly,Marks_train),3))
```

```python
C= 1
logit = LogisticRegression(C=C, random_state=17)
logit.fit(X_poly, y =Marks_train)

plot_boundary(logit,Data_train['blue'],Data_train['red'],grid_step=.01, poly_featurizer= poly)

plt.scatter(Data_train[Marks_train== 0][['blue']],Data_train[Marks_train== 0][['red']])
plt.scatter(Data_train[Marks_train== 1][['blue']],Data_train[Marks_train== 1][['red']], c='Orange')
plt.title('Logit with C=%s' % C)

print("Accuracy on training set:", round(logit.score(X_poly,Marks_train),3))
```

```python
C= 1e5
logit = LogisticRegression(C=C, random_state=17)
logit.fit(X_poly, y =Marks_train)

plot_boundary(logit,Data_train['blue'],Data_train['red'],grid_step=.01, poly_featurizer= poly)

plt.scatter(Data_train[Marks_train== 0][['blue']],Data_train[Marks_train== 0][['red']])
plt.scatter(Data_train[Marks_train== 1][['blue']],Data_train[Marks_train== 1][['red']], c='Orange')
plt.title('Logit with C=%s' % C)

print("Accuracy on training set:", round(logit.score(X_poly,Marks_train),3))
```

```python
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=17)
c_values = np.logspace(-2,3,500)
logit_searcher = LogisticRegressionCV(Cs=c_values,cv=skf, verbose=1, n_jobs=-1)
logit_searcher.fit(X_poly, y =Marks_train)
```

```python
plt.plot(c_values, np.mean(logit_searcher.scores_[1], axis=0))
plt.xlabel('C')
plt.ylabel('Mean CV-accuracy')
```
