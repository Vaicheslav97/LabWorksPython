```python
import pandas as pd
%config InlineBackend.figure_format = 'png'
df = pd.read_csv("loan_data_set.csv")
import seaborn as sns 


dfk = df[df['ApplicantIncome']!=' '][['Dependents','Self_Employed','ApplicantIncome','CoapplicantIncome','Property_Area','LoanAmount','Loan_Status','Education','Loan_Amount_Term','Marrie']]
dfk.dropna()
 
import math
Pdict = {hash("Urban"): 0, hash("Semiurban"): 1, hash("Rural"): 2}
dfk['Property_Area']=dfk['Property_Area'].apply(lambda x: hash(x))

dfk['Property_Area']=dfk['Property_Area'].apply(lambda x: Pdict[x]) 

dfk['Dependents']=dfk['Dependents'].apply(lambda x: -2 if x==''  else x) 
dfk['Dependents']=dfk['Dependents'].apply(lambda x: -1 if x=="3+" else float(x) )
dfk['Dependents']=dfk['Dependents'].apply(lambda x: -2 if math.isnan(x)  else x) 

dfk['Marrie']=dfk['Marrie'].apply(lambda x: 1 if x=='Yes' else 0)
dfk['Self_Employed']=dfk['Self_Employed'].apply(lambda x: 1 if x=='Yes' else 0)
dfk['Education']=dfk['Education'].apply(lambda x: 1 if x=='Graduate' else 0)

dfk['LoanAmount']=dfk['LoanAmount'].apply(lambda x: 0 if math.isnan(x) else x)
dfk['ApplicantIncome']=dfk['ApplicantIncome'].apply(lambda x: 0 if math.isnan(x) else x/100)
dfk['CoapplicantIncome']=dfk['CoapplicantIncome'].apply(lambda x: 0 if math.isnan(x) else x/100)
dfk['FIncome']=dfk['ApplicantIncome']+dfk['CoapplicantIncome']
dfk['FIncome']=dfk['FIncome'].apply(lambda x: float (x/100)  )
dfk['LoanAmount']=dfk['LoanAmount'].apply(lambda x: float (x/100)) 
dfk['LAPT']=dfk['LoanAmount']/dfk['Loan_Amount_Term']
 

#dfk['HaveLoan']=dfk['LoanAmount'].apply(lambda x: 0 if x==0 else 1)
dfk['Loan_Status']=dfk['Loan_Status'].apply(lambda x: 0 if x=='N' else 1)
dfkh= dfk[dfk['ApplicantIncome']!=''][['Dependents','FIncome','Marrie']]
dfkid = dfk[dfk['ApplicantIncome']!=''][['LoanAmount','Dependents']]#
dfkmd = dfk['Loan_Status']
#dfkid
#dfkmd
#df.head()
#df
dfk.head()
#df.info()
```

```python
def plot_boundary(clf, X, Y, grid_step=.01, poly_featurizer=None):
    x_min, x_max= X.min()-.1, X.max()+.1
    y_min, y_max= Y.min()-.1, Y.max()+.1
    xx,yy = np.meshgrid(np.arange(x_min, x_max, grid_step),
                        np.arange(y_min, y_max, grid_step))
    Z =clf.predict(poly_featurizer.transform(np.c_[xx.ravel(),yy.ravel()]))
    Z = Z.reshape(xx.shape)
   # plt.contour(xx,yy,Z,cmap=plt.cm.paired)
    plt.contour(xx,yy,Z,cmap=plt.cm.get_cmap("Spectral"))
```
